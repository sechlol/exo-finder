```python
from exo_finder.training.data_loading.lightcurve_dataset import TorchLightcurveDataset
from exo_finder.constants import HDF5_KEY_LC_DATA, HDF5_KEY_LC_SIZES, HDF5_KEY_LC_TIC_IDS
from torch.utils.data import DataLoader, RandomSampler, BatchSampler
```

```python
dataset = TorchLightcurveDataset(keys_to_fetch=[HDF5_KEY_LC_DATA, HDF5_KEY_LC_SIZES, HDF5_KEY_LC_TIC_IDS])
```

```python
sampler = BatchSampler(
    RandomSampler(dataset),
    batch_size=100,
    drop_last=False)

loader = DataLoader(dataset, sampler=sampler)

for i, data in enumerate(loader):
    print(data["lc_tic_ids"].squeeze())
    if i == 3:
        break
```
